### Question 1: Some of the orders are stored in another csv file named `bigkart_newsales`. Read the csv file, store it in a data frame and add it to the `bigkart_sales` data frame. Find the total sales value of the category 'Office Supplies' after combining the dataframes

```python
bigkart_sales = pd.read_csv("AssignData2/bigkart_sales.csv", header = 0)
BkNew_sales = pd.read_csv("AssignData2/bigkart_newsales.csv", header = 0)
```

```python
concat = [bigkart_sales, BkNew_sales]
all_sales = pd.concat(concat,axis=0,ignore_index=True)
```

```python
all_sales.head()
```

```python
all_sales.columns
```

```python
A1 = all_sales.iloc[np.where((all_sales["Category"] == "Office Supplies")
                       )].Sales.sum()

print("Answare 1 :\nTotal sales value of the category 'Office Supplies' after combining the dataframes =", A1)
```

### Question 2: There are some duplicate rows in the data frame. Drop these rows and calculate the total sales value of the category Office Supplies.


```python
all_sales.drop_duplicates(subset=None,keep='first',inplace = True)
```

```python
A2 = all_sales.iloc[np.where((all_sales["Category"] == "Office Supplies")
                       )].Sales.sum()

print("Answare 2 :\nTotal sales value of the category Office Supplies after removing the Duplicate rows =", A2)
```

### Question 3: Find the most profitable category and sub category combination based on the net profit.

#### Answare 3: Based on the net profit, the most profitable category is "Technology" and sub category is "Phones" combination.

### Question 4: How many invalid order IDs are there in the data frame. An order id is of the form AZ-2011-Y where Y represents a whole number. A Order ID is said to be valid only if Y consists of 7 digits. Find the number of invalid order order IDs in the data frame.


```python
all_sales['Order No'] = all_sales['Order ID'].str.split('-')
all_sales['Order No'] = all_sales['Order No'].apply(lambda x: x[2])
```

```python
all_sales['Order No']
```

```python
all_sales["DCheck"] = all_sales['Order No'].apply(lambda x: len(x))
```

```python
all_sales.head()
```

```python
all_sales.groupby(['Order ID','Order No'])["DCheck"].agg("value_counts")
```

```python
all_sales['DCheck'].value_counts()
```

#### Answare 4:
The number of invalid order order IDs in the data frame = 7
